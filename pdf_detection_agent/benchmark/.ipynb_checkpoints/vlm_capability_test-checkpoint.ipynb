```python
%matplotlib inline
import sys, os, time
from concurrent.futures import ThreadPoolExecutor
sys.path.insert(0, ".")
from utils import (
    extract_gt_boxes, extract_bboxes, evaluate, get_tp_boxes,
    dedup, run_inference, build_tile, P_BASE, P_ANCHOR,
    ollama_chat_with_timeout, validate_json,
)
from PIL import Image as _Image, ImageDraw
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

ANNOT_PATH = "/home/jiezhi/tmp/test_floorplan_tile.png"
CLEAN_PATH = "/tmp/test_floorplan_tile.png"
OUT_DIR    = "/tmp/vlm_capability_test"
os.makedirs(OUT_DIR, exist_ok=True)

TILE_PATH    = build_tile(CLEAN_PATH, 1024)   # normal runs
TILE_PATH_2K = build_tile(CLEAN_PATH, 2048)   # LOUD: 4x more vision patches for the encoder
GT           = extract_gt_boxes(ANNOT_PATH, 1024)
SEEDS        = [GT[int(i * (len(GT)-1) / 9)] for i in range(10)]

gt_img  = _Image.open(TILE_PATH).convert("RGB")
gt_draw = ImageDraw.Draw(gt_img)
for g in GT:
    gt_draw.rectangle(g, outline=(0, 210, 0), width=3)
for s in SEEDS:
    gt_draw.rectangle(s, outline=(0, 210, 220), width=4)
gt_draw.rectangle([0, 0, 1024, 22], fill=(0, 0, 0))
gt_draw.text((4, 4), f"GROUND TRUTH  {len(GT)} columns (GREEN)   10 injection seeds (CYAN)", fill=(255, 230, 0))
gt_draw.rectangle([0, 1002, 1024, 1024], fill=(20, 20, 20))
gt_draw.text((4, 1006), "GREEN = all 86 GT columns   CYAN = 10 spread seeds used in ANCHOR rounds", fill=(180, 180, 180))
gt_path = f"{OUT_DIR}/ground_truth.png"
gt_img.save(gt_path)

fig, ax = plt.subplots(figsize=(10, 10))
ax.imshow(mpimg.imread(gt_path))
ax.set_title(f"Ground Truth  {len(GT)} GT columns (GREEN) | 10 injection seeds (CYAN)", fontsize=12)
ax.axis("off")
plt.tight_layout()
plt.show()
print(f"GT: {len(GT)} columns | output: {OUT_DIR}")


def annotate(dets, pool, label, fname, tile=None):
    img  = _Image.open(tile or TILE_PATH).convert("RGB").resize((1024, 1024))
    draw = ImageDraw.Draw(img)
    for b in pool:
        draw.rectangle(b, outline=(0, 210, 220), width=2)
    for b in extract_bboxes(dets):
        draw.rectangle(b, outline=(230, 30, 30), width=3)
    m = evaluate(dets, GT)
    hdr = (f"{label}  "
           f"dets={len(extract_bboxes(dets))} TP={m['tp']} FP={m['fp']} FN={m['fn']} "
           f"P={m['p']:.2f} R={m['r']:.2f} F1={m['f1']:.2f}  pool={len(pool)}")
    draw.rectangle([0, 0, 1024, 22], fill=(0, 0, 0))
    draw.text((4, 4), hdr, fill=(255, 230, 0))
    draw.rectangle([0, 1002, 1024, 1024], fill=(20, 20, 20))
    draw.text((4, 1006), "CYAN=injected pool   RED=model detections", fill=(180, 180, 180))
    path = f"{OUT_DIR}/{fname}"
    img.save(path)
    return path, m


def run(tag, timeout=300, rounds=5):
    safe  = tag.replace(":", "_").replace(".", "_").replace("/", "_")
    pool  = []
    paths = []
    titles = []
    print(f"\n{'='*60}\n{tag}\n{'='*60}")
    for rnd in range(1, rounds + 1):
        if rnd == 1:
            label, prompt, pool_before = "R1 BASELINE", P_BASE, []
        elif rnd == 2:
            pool        = dedup(list(SEEDS))
            label       = "R2 ANCHOR(seed=10)"
            prompt      = P_ANCHOR(pool)
            pool_before = pool
        else:
            label       = f"R{rnd} ANCHOR(pool={len(pool)})"
            prompt      = P_ANCHOR(pool)
            pool_before = pool
        print(f"  [{rnd}/{rounds}] {label} ...", end=" ", flush=True)
        r = run_inference(tag, prompt, TILE_PATH, timeout_s=timeout)
        m = evaluate(r["dets"], GT)
        print(f"dets={r['n_dets']}  TP={m['tp']}  FP={m['fp']}  F1={m['f1']:.3f}  ({r['elapsed']}s)")
        path, _ = annotate(r["dets"], pool_before, label, f"{safe}_r{rnd}.png")
        paths.append(path)
        titles.append(label)
        if rnd >= 2:
            new_tps = get_tp_boxes(r["dets"], GT)
            pool    = dedup(pool + new_tps)
            if new_tps:
                print(f"    pool +{len(new_tps)} TPs -> pool={len(pool)}")
    fig, axes = plt.subplots(2, 3, figsize=(18, 12))
    for i, (path, title) in enumerate(zip(paths, titles)):
        ax = axes[i // 3][i % 3]
        ax.imshow(mpimg.imread(path))
        ax.set_title(title, fontsize=10, fontweight="bold")
        ax.axis("off")
    axes[1][2].set_visible(False)
    fig.suptitle(tag, fontsize=14, fontweight="bold", y=1.01)
    plt.tight_layout()
    plt.show()


# STRESS prompt: forces 3000-8000 tokens instead of ~100 for short JSON
STRESS_PROMPT = (
    "You are examining a 2048x2048 architectural floor plan image. "
    "Your task is to identify and describe EVERY structural column in exhaustive detail.\n\n"
    "For each column you find, write a paragraph covering:\n"
    "  1. Exact pixel bounding box (x1, y1, x2, y2) within the 2048x2048 image\n"
    "  2. Shape: square / rectangle / round / i-beam\n"
    "  3. Approximate width and height in pixels\n"
    "  4. Fill style: solid filled / hatched / outlined only\n"
    "  5. Nearby grid axis labels (letters A B C or numbers 1 2 3)\n"
    "  6. Adjacent wall lines and their orientation\n"
    "  7. Your confidence that this is a structural column (0.0 to 1.0)\n\n"
    "After your prose description of ALL columns, output a complete JSON list:\n"
    '[{"type":"column","bbox":[x1,y1,x2,y2],"confidence":0.95}]\n\n'
    "Be thorough. There are approximately 80-100 columns in this image. "
    "Scan every grid intersection systematically from top-left to bottom-right. "
    "Do not stop early. Describe every single column you find before outputting JSON. "
    "This is a structural safety audit -- completeness is critical."
)


def _one_call(tag, prompt, tile, timeout, opts):
    """Single inference call - runs inside a worker thread."""
    t0 = time.time()
    try:
        resp = ollama_chat_with_timeout(
            timeout,
            model=tag,
            messages=[{"role": "user", "content": prompt, "images": [tile]}],
            options=opts,
            keep_alive=-1,  # LOUD: keep model in VRAM between calls
        )
        raw = resp["message"]["content"] if resp else ""
        err = None
    except Exception as e:
        raw = ""; err = str(e)
    elapsed = round(time.time() - t0, 1)
    ok, parsed, _ = validate_json(raw)
    dets = parsed if (ok and isinstance(parsed, list)) else []
    return dets, elapsed, err


def run_turbo(tag, rounds=20, concurrency=2, timeout=480):
    safe  = tag.replace(":", "_").replace(".", "_").replace("/", "_")
    pool  = dedup(list(SEEDS))
    paths = []
    titles = []
    opts = {
        "num_predict": 8192,   # LOUD: max generation -- GPU stays hot for the full output
        "num_ctx":     32768,  # LOUD: huge KV cache -- saturates VRAM even on small models
        "num_batch":   512,    # LOUD: large prefill batch -- spikes GPU at each call start
        "temperature": 0.5,    # LOUD: higher temp = less early stopping = more tokens
        "repeat_penalty": 1.0, # LOUD: no repetition penalty = model generates freely
    }
    print(f"\n{'='*60}")
    print(f"{tag}  [TURBO {rounds} rounds x{concurrency} concurrent]")
    print(f"{'='*60}")
    for rnd in range(1, rounds + 1):
        label       = "R1 BASELINE" if rnd == 1 else f"R{rnd} ANCHOR(pool={len(pool)})"
        # LOUD: STRESS_PROMPT for round 1 forces 3000-8000 tokens vs ~100 for JSON
        prompt      = STRESS_PROMPT if rnd == 1 else P_ANCHOR(pool)
        pool_before = []            if rnd == 1 else pool
        print(f"  [{rnd:>2}/{rounds}] {label} -- firing {concurrency} concurrent calls...",
              end=" ", flush=True)
        t_rnd = time.time()
        # LOUD: fire concurrency=2 calls simultaneously -- GPU handles multiple streams
        #       at once; fan spins up to sustain throughput across both
        with ThreadPoolExecutor(max_workers=concurrency) as ex:
            futures = [
                ex.submit(_one_call, tag, prompt, TILE_PATH_2K, timeout, opts)
                for _ in range(concurrency)
            ]
            results = [f.result() for f in futures]
        elapsed_rnd = round(time.time() - t_rnd, 1)
        # Merge + dedup detections from all concurrent calls
        all_dets = []
        for dets, elapsed, err in results:
            all_dets.extend(dets)
            if err:
                print(f"\n    ! {err}")
        all_dets = dedup(all_dets)
        m = evaluate(all_dets, GT)
        print(f"merged_dets={len(all_dets)}  TP={m['tp']}  FP={m['fp']}  "
              f"F1={m['f1']:.3f}  ({elapsed_rnd}s total)")
        path, _ = annotate(all_dets, pool_before, label,
                           f"{safe}_turbo_r{rnd}.png", tile=TILE_PATH_2K)
        paths.append(path)
        titles.append(f"R{rnd}")
        new_tps = get_tp_boxes(all_dets, GT)
        pool = dedup(pool + new_tps)
        if new_tps:
            print(f"    pool +{len(new_tps)} TPs -> pool={len(pool)}")
    fig, axes = plt.subplots(2, 3, figsize=(18, 12))
    for i, (path, title) in enumerate(zip(paths[-5:], titles[-5:])):
        ax = axes[i // 3][i % 3]
        ax.imshow(mpimg.imread(path))
        ax.set_title(title, fontsize=10, fontweight="bold")
        ax.axis("off")
    axes[1][2].set_visible(False)
    fig.suptitle(f"{tag} -- last 5 of {rounds} rounds", fontsize=13, fontweight="bold")
    plt.tight_layout()
    plt.show()


print("Ready.")
```

```python
# LOUD: 20 rounds x 2 concurrent calls = 40 total inference calls back-to-back
# LOUD: round 1 uses STRESS_PROMPT -- forces 3000-8000 tokens (vs ~100 for JSON)
# LOUD: 2048px image -- 4x more vision patches than 1024px
# LOUD: num_ctx=32768 -- huge KV cache fills VRAM
# LOUD: num_batch=512 -- large prefill batch spikes GPU at each call start
# LOUD: num_predict=8192 -- max output length keeps GPU busy for full token run
# LOUD: temperature=0.5 -- reduces early stopping so model keeps generating
# LOUD: keep_alive=-1 -- model stays resident in VRAM with no unload gaps
# LOUD: concurrency=2 -- two inference streams running simultaneously
# To quieten: lower rounds/concurrency, switch TILE_PATH_2K->TILE_PATH, reduce num_ctx
run_turbo("gemma3:4b", rounds=20, concurrency=2)
```
