# Ollama Vision Model Evaluation — Column Detection Benchmark

Systematically pulls and tests every Ollama vision model ≤ 6 GB against a labelled floor plan PDF.
Produces a ranked table so you can pick the best local model before spending on GPT-4o credits.

**Ground truth**: 92 true columns (67 confirmed + 25 human-added) from `detections.db` corrections.  
**Known false-positives**: 197 boxes that SEA-LION reported but humans rejected.  
**Scoring**: 6-point rubric — valid JSON, has bboxes, count in range, P > 0.2, R > 0.3, F1 > 0.25.  
**Pass threshold**: score ≥ 4/6.

```python
import sys, os, json, time, subprocess, re, math, shutil
from pathlib import Path
from collections import defaultdict, Counter
from datetime import datetime
import urllib.request, urllib.error

import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from PIL import Image, ImageDraw
import numpy as np

AGENT_DIR = Path(".").resolve()
sys.path.insert(0, str(AGENT_DIR))
import agent

print(f"Agent dir    : {AGENT_DIR}")
print(f"TILE_SIZE    : {agent.TILE_SIZE}")
print(f"RENDER_DPI   : {agent.RENDER_DPI}")
```

## 1 — Configuration

```python
OLLAMA_URL  = os.getenv("OLLAMA_URL", "http://localhost:11434")
TEST_PDF    = Path("/home/jiezhi/Documents/floor-plan-pdf/2aad5b45-ce32-40a6-a00d-e5b0e67f5d08.pdf")
LOG_FILE    = AGENT_DIR / "vision_model_evaluation_log.json"
IOU_THRESH  = 0.25   # IoU threshold for TP matching
CONF_THRESH = 0.3    # minimum confidence to count a detection
PASS_SCORE  = 4      # minimum score to be considered "suitable"

# Models to evaluate — (name, approx_gb, notes)
# Ordered small → large; already-pulled first to get baseline before downloading
MODELS_TO_TEST = [
    ("moondream:latest",                         1.7,  "Already pulled — near-zero baseline"),
    ("granite3.2-vision:2b",                     1.9,  "IBM Granite 2B vision"),
    ("llava-phi3:3.8b",                          2.9,  "LLaVA on Phi-3 mini"),
    ("qwen2.5vl:3b",                             3.2,  "Already pulled — Qwen 3B"),
    ("aisingapore/Gemma-SEA-LION-v4-4B-VL:latest", 3.3, "Already pulled — SEA-LION 4B"),
    ("llava:7b",                                 4.7,  "Original LLaVA 7B"),
    ("minicpm-v:8b",                             5.5,  "MiniCPM-V 8B (strong OCR+vision)"),
    ("llava-llama3:8b",                          5.5,  "LLaVA on LLaMA-3"),
    ("qwen2.5vl:7b",                             6.0,  "Qwen2.5-VL 7B — best small VL model"),
]

print(f"Test PDF     : {TEST_PDF.name}  exists={TEST_PDF.exists()}")
print(f"Models queue : {len(MODELS_TO_TEST)} models ({sum(g for _,g,_ in MODELS_TO_TEST):.0f} GB total if all pulled)")
print(f"Log file     : {LOG_FILE}")
```

## 2 — Ollama Helpers

```python
def ollama_running() -> bool:
    try:
        with urllib.request.urlopen(f"{OLLAMA_URL}/api/tags", timeout=3):
            return True
    except Exception:
        return False


def ollama_pulled_models() -> set:
    try:
        with urllib.request.urlopen(f"{OLLAMA_URL}/api/tags", timeout=5) as r:
            data = json.loads(r.read())
        return {m["name"] for m in data.get("models", [])}
    except Exception:
        return set()


def ollama_pull(model: str) -> bool:
    """Pull model via CLI. Returns True on success."""
    print(f"  Pulling {model} ...", flush=True)
    result = subprocess.run(
        ["ollama", "pull", model],
        capture_output=False,
        timeout=1800   # 30 min max
    )
    return result.returncode == 0


def ollama_chat(model: str, prompt: str, image_b64: str,
                timeout: int = 120) -> tuple[str, float]:
    """
    Call Ollama /api/chat with one image.
    Returns (response_text, elapsed_seconds).
    """
    payload = json.dumps({
        "model": model,
        "messages": [{
            "role": "user",
            "content": prompt,
            "images": [image_b64]
        }],
        "stream": False,
        "options": {"temperature": 0.1, "num_predict": 2048}
    }).encode()

    req = urllib.request.Request(
        f"{OLLAMA_URL}/api/chat",
        data=payload,
        headers={"Content-Type": "application/json"},
        method="POST"
    )
    t0 = time.time()
    with urllib.request.urlopen(req, timeout=timeout) as r:
        data = json.loads(r.read())
    elapsed = time.time() - t0
    text = data.get("message", {}).get("content", "")
    return text, elapsed


# Verify Ollama is up
if not ollama_running():
    print("Ollama not running — starting...")
    subprocess.Popen(["ollama", "serve"], stdout=subprocess.DEVNULL, stderr=subprocess.DEVNULL)
    time.sleep(4)

pulled = ollama_pulled_models()
print(f"Ollama running: {ollama_running()}")
print(f"Already pulled: {sorted(pulled)}")
```

## 3 — Detection Prompt & Parser

```python
# A single prompt that works across model families.
# Shorter than the GPT-4o prompt — smaller models follow shorter instructions better.
_DETECT_PROMPT = """\
You are analysing an architectural floor plan tile ({sz}x{sz} px).
Find all structural columns — the solid/filled/outlined/hatched squares, rectangles, or circles
at beam or grid intersections. These are small load-bearing shapes, often labelled C1, C2, H-C1.

DO NOT report:
- Grid balloons (open circles at grid line ends with letters/numbers A, B, 1, 2)
- Wall lines, door arcs, dimension arrows, text, north arrows

For each column found, output JSON:
{{"columns": [
  {{"bbox": [x1, y1, x2, y2], "confidence": 0.90, "notes": "solid square"}}
]}}
bbox coordinates are pixel positions within this {sz}x{sz} tile.
If none found: {{"columns": []}}
Respond with JSON only."""


def detect_tile_ollama(tile: Image.Image, info, model: str,
                       max_retries: int = 3) -> tuple[list[dict], float, bool]:
    """
    Detect columns in one tile using an Ollama model.
    Returns (detections, elapsed_seconds, valid_json).
    """
    content = tile.crop((0, 0, info.width, info.height))
    send    = content.resize((agent.VISION_SIZE, agent.VISION_SIZE), Image.LANCZOS)
    b64     = agent._pil_to_b64(send)
    prompt  = _DETECT_PROMPT.format(sz=agent.VISION_SIZE)

    for attempt in range(max_retries):
        try:
            raw, elapsed = ollama_chat(model, prompt, b64)
            parsed = agent._parse_json(raw)
            valid_json = bool(parsed.get("columns") is not None)
            dets = []
            for col in parsed.get("columns", []):
                bbox = col.get("bbox", [])
                if len(bbox) != 4:
                    continue
                if not all(isinstance(v, (int, float)) for v in bbox):
                    continue
                # Handle normalised 0-1 coords
                if all(0.0 <= v <= 1.0 for v in bbox):
                    bbox = [v * agent.VISION_SIZE for v in bbox]
                sx = info.width  / agent.VISION_SIZE
                sy = info.height / agent.VISION_SIZE
                bbox = [bbox[0]*sx, bbox[1]*sy, bbox[2]*sx, bbox[3]*sy]
                try:
                    conf = float(col.get("confidence", 0.5))
                    conf = conf / 100.0 if conf > 1.0 else conf
                    conf = max(0.0, min(1.0, conf))
                except (TypeError, ValueError):
                    conf = 0.5
                dets.append({
                    "bbox_tile": bbox,
                    "bbox_page": agent._to_page_bbox(bbox, info),
                    "confidence": conf,
                    "notes": col.get("notes", ""),
                    "tile_hash": agent._tile_hash(send),
                    "tile_index": info.index, "page_num": info.page_num,
                    "shape": "column"
                })
            return dets, elapsed, valid_json
        except Exception as e:
            if attempt < max_retries - 1:
                time.sleep(2)
            else:
                return [], 0.0, False
    return [], 0.0, False


def detect_pdf_ollama(pdf_path: Path, model: str, page_num: int = 0) -> dict:
    """Run full-PDF detection with an Ollama model. Returns result dict."""
    img, n_pages = agent._pdf_to_pil(pdf_path, page_num, agent.RENDER_DPI)
    W, H = img.size
    tiles = list(agent._tiles(img, page_num))
    all_dets, skipped, total_time = [], 0, 0.0
    valid_json_count = 0

    for tile_img, tile_info in tiles:
        if not agent._tile_has_content(tile_img, tile_info):
            skipped += 1
            continue
        dets, elapsed, valid_json = detect_tile_ollama(tile_img, tile_info, model)
        total_time += elapsed
        if valid_json:
            valid_json_count += 1
        all_dets.extend(dets)

    final = agent._nms(all_dets)
    final = agent._cell_dedup(final, W, H, "bbox_page", 12, 12)
    processed = len(tiles) - skipped

    return {
        "model": model, "detections": final,
        "total_columns": len(final),
        "tiles": len(tiles), "tiles_skipped": skipped,
        "valid_json_tiles": valid_json_count,
        "total_time_s": round(total_time, 1),
        "avg_time_per_tile_s": round(total_time / processed, 1) if processed else 0,
        "avg_confidence": round(sum(d["confidence"] for d in final)/len(final), 3) if final else 0.0,
    }

print("Detection functions ready.")
```

## 4 — Ground Truth & Scoring

```python
def load_ground_truth(file_id: str, page_width: int, page_height: int) -> dict:
    con = agent._db()
    rows = con.execute(
        "SELECT action, tile_index, x_offset, y_offset, "
        "       bbox_x1, bbox_y1, bbox_x2, bbox_y2"
        " FROM corrections WHERE file_path LIKE ?",
        (f"%{file_id}%",)
    ).fetchall()
    con.close()
    result = {"tp": [], "fp": [], "fn": []}
    for r in rows:
        x_off, y_off = r["x_offset"], r["y_offset"]
        tw = min(agent.TILE_SIZE, page_width  - x_off)
        th = min(agent.TILE_SIZE, page_height - y_off)
        bbox = [r["bbox_x1"]*tw+x_off, r["bbox_y1"]*th+y_off,
                r["bbox_x2"]*tw+x_off, r["bbox_y2"]*th+y_off]
        entry = {"bbox": bbox}
        if   r["action"] == "confirm": result["tp"].append(entry)
        elif r["action"] == "reject":  result["fp"].append(entry)
        else:                          result["fn"].append(entry)
    return result


def iou(a, b):
    ix1,iy1 = max(a[0],b[0]), max(a[1],b[1])
    ix2,iy2 = min(a[2],b[2]), min(a[3],b[3])
    inter = max(0, ix2-ix1) * max(0, iy2-iy1)
    if inter == 0: return 0.0
    return inter / ((a[2]-a[0])*(a[3]-a[1]) + (b[2]-b[0])*(b[3]-b[1]) - inter)


def evaluate(detections: list[dict], gt: dict,
             iou_thresh=IOU_THRESH, conf_thresh=CONF_THRESH) -> dict:
    gt_boxes = [g["bbox"] for g in gt["tp"]] + [g["bbox"] for g in gt["fn"]]
    fp_boxes = [g["bbox"] for g in gt["fp"]]
    filtered = [d for d in detections if d.get("confidence", 0) >= conf_thresh]
    matched  = set()
    tp = 0
    for det in sorted(filtered, key=lambda d: d["confidence"], reverse=True):
        db = det["bbox_page"]
        best_i, best_v = -1, 0.0
        for i, gb in enumerate(gt_boxes):
            if i in matched: continue
            v = iou(db, gb)
            if v > best_v: best_v, best_i = v, i
        if best_v >= iou_thresh:
            matched.add(best_i); tp += 1
    fp_count = len(filtered) - tp
    fn_count = len(gt_boxes) - tp
    p = tp / len(filtered) if filtered else 0.0
    r = tp / len(gt_boxes)  if gt_boxes  else 0.0
    f = 2*p*r/(p+r) if (p+r) > 0 else 0.0
    return {"tp": tp, "fp": fp_count, "fn": fn_count,
            "precision": round(p,3), "recall": round(r,3), "f1": round(f,3),
            "n_det": len(filtered), "n_gt": len(gt_boxes)}


def score_model(run_result: dict, metrics: dict, gt: dict) -> tuple[int, list[str]]:
    """
    6-point rubric:
      1. Valid JSON on ≥ 50% of tiles
      2. Has bbox coordinates (any detection at all)
      3. Detection count in plausible range (5 – 200)
      4. Precision > 0.20
      5. Recall > 0.30
      6. F1 > 0.25
    """
    n_proc = run_result["tiles"] - run_result["tiles_skipped"]
    points = []
    score  = 0

    # 1 — valid JSON
    json_rate = run_result["valid_json_tiles"] / n_proc if n_proc else 0
    if json_rate >= 0.5:
        score += 1; points.append(f"✓ Valid JSON {json_rate:.0%} of tiles")
    else:
        points.append(f"✗ Poor JSON rate ({json_rate:.0%})")

    # 2 — has detections with bboxes
    if run_result["total_columns"] > 0:
        score += 1; points.append(f"✓ Detects columns ({run_result['total_columns']} found)")
    else:
        points.append("✗ Zero detections")

    # 3 — plausible count
    n = run_result["total_columns"]
    if 5 <= n <= 200:
        score += 1; points.append(f"✓ Count in range ({n})")
    else:
        points.append(f"✗ Count out of range ({n})")

    # 4 — precision
    if metrics["precision"] > 0.20:
        score += 1; points.append(f"✓ Precision {metrics['precision']:.2f} > 0.20")
    else:
        points.append(f"✗ Precision {metrics['precision']:.2f} ≤ 0.20")

    # 5 — recall
    if metrics["recall"] > 0.30:
        score += 1; points.append(f"✓ Recall {metrics['recall']:.2f} > 0.30")
    else:
        points.append(f"✗ Recall {metrics['recall']:.2f} ≤ 0.30")

    # 6 — F1
    if metrics["f1"] > 0.25:
        score += 1; points.append(f"✓ F1 {metrics['f1']:.2f} > 0.25")
    else:
        points.append(f"✗ F1 {metrics['f1']:.2f} ≤ 0.25")

    return score, points


# Pre-load ground truth
gt_img, _ = agent._pdf_to_pil(TEST_PDF, 0, agent.RENDER_DPI)
PW, PH    = gt_img.size
GT        = load_ground_truth("2aad5b45", PW, PH)
GT_TOTAL  = len(GT["tp"]) + len(GT["fn"])
print(f"Ground truth loaded: {GT_TOTAL} true columns  ({len(GT['tp'])} confirmed + {len(GT['fn'])} adds)")
print(f"Known false-positive pool: {len(GT['fp'])} boxes")
```

## 5 — Main Evaluation Loop

Iterates through every model: pulls if needed, detects, scores, logs.

```python
# Load existing log (resume if interrupted)
if LOG_FILE.exists():
    log_entries = json.loads(LOG_FILE.read_text())
    already_tested = {e["model"] for e in log_entries}
    print(f"Resuming — {len(already_tested)} models already in log: {already_tested}")
else:
    log_entries, already_tested = [], set()
    print("Starting fresh evaluation.")

print()
```

```python
for model_name, approx_gb, notes in MODELS_TO_TEST:

    if model_name in already_tested:
        print(f"[SKIP] {model_name} — already in log")
        continue

    print(f"\n{'='*65}")
    print(f"MODEL : {model_name}")
    print(f"SIZE  : ~{approx_gb:.1f} GB   | {notes}")
    print(f"{'='*65}")

    entry = {
        "model": model_name, "approx_gb": approx_gb, "notes": notes,
        "status": "failed", "score": 0,
        "tested_at": datetime.now().isoformat()
    }

    # ── Step 1: Pull if needed ────────────────────────────────────────────────
    pulled_now = ollama_pulled_models()
    model_base = model_name.split(":")[0]
    is_pulled  = any(model_base in m for m in pulled_now)

    if not is_pulled:
        print(f"  Pulling {model_name} (~{approx_gb:.1f} GB) ...")
        t_pull = time.time()
        ok = ollama_pull(model_name)
        pull_time = round(time.time() - t_pull, 1)
        if not ok:
            print(f"  PULL FAILED — skipping")
            entry["status"] = "pull_failed"
            log_entries.append(entry)
            LOG_FILE.write_text(json.dumps(log_entries, indent=2))
            already_tested.add(model_name)
            continue
        print(f"  Pulled in {pull_time:.0f}s")
        entry["pull_time_s"] = pull_time
    else:
        print(f"  Already pulled.")

    # ── Step 2: Detect ────────────────────────────────────────────────────────
    fail_count = 0
    run_result = None
    for attempt in range(3):
        try:
            print(f"  Detecting (attempt {attempt+1}/3) ...", flush=True)
            run_result = detect_pdf_ollama(TEST_PDF, model_name)
            print(f"  → {run_result['total_columns']} columns | "
                  f"{run_result['total_time_s']:.0f}s total | "
                  f"valid_json={run_result['valid_json_tiles']}/{run_result['tiles']-run_result['tiles_skipped']} tiles")
            break
        except Exception as e:
            fail_count += 1
            print(f"  ERROR attempt {attempt+1}: {e}")
            time.sleep(3)

    if run_result is None:
        print("  3 consecutive failures — marking as failed")
        entry["status"] = "detection_failed"
        log_entries.append(entry)
        LOG_FILE.write_text(json.dumps(log_entries, indent=2))
        already_tested.add(model_name)
        continue

    # ── Step 3: Score ─────────────────────────────────────────────────────────
    metrics = evaluate(run_result["detections"], GT)
    score, points = score_model(run_result, metrics, GT)

    print(f"  Metrics : P={metrics['precision']:.3f}  R={metrics['recall']:.3f}  F1={metrics['f1']:.3f}")
    print(f"  Score   : {score}/6  {'PASS ✓' if score >= PASS_SCORE else 'FAIL ✗'}")
    for pt in points:
        print(f"    {pt}")

    # ── Step 4: Log ───────────────────────────────────────────────────────────
    entry.update({
        "status": "pass" if score >= PASS_SCORE else "fail",
        "score": score,
        "total_columns": run_result["total_columns"],
        "precision": metrics["precision"],
        "recall": metrics["recall"],
        "f1": metrics["f1"],
        "avg_confidence": run_result["avg_confidence"],
        "valid_json_tiles": run_result["valid_json_tiles"],
        "response_time_seconds": run_result["total_time_s"],
        "avg_time_per_tile_s": run_result["avg_time_per_tile_s"],
        "rubric": points,
    })
    log_entries.append(entry)
    LOG_FILE.write_text(json.dumps(log_entries, indent=2))
    already_tested.add(model_name)

print(f"\n{'='*65}")
print(f"All models tested. Log saved to {LOG_FILE}")
```

## 6 — Results Table

```python
# Load log and sort by score desc, then F1 desc
all_entries = json.loads(LOG_FILE.read_text())
all_entries.sort(key=lambda x: (x.get("score", 0), x.get("f1", 0)), reverse=True)

# Add GPT-4o SEA-LION historical baselines from detections.db
BASELINES = [
    {"model": "gpt-4o [QUOTA EXCEEDED]", "approx_gb": 0, "status": "error",
     "score": "N/A", "total_columns": 0, "precision": None, "recall": None,
     "f1": None, "avg_confidence": None, "response_time_seconds": None},
    {"model": "SEA-LION 4B (est. from corrections)", "approx_gb": 3.3,
     "status": "fail", "score": 3,
     "total_columns": 75, "precision": 0.254, "recall": 0.728, "f1": 0.376,
     "avg_confidence": 0.952, "response_time_seconds": None},
]

display_entries = all_entries + BASELINES
display_entries.sort(key=lambda x: (x.get("score") if isinstance(x.get("score"), int) else -1,
                                     x.get("f1") or 0), reverse=True)

# Print table
print(f"\n{'='*100}")
print(f"VISION MODEL EVALUATION RESULTS — {TEST_PDF.name}")
print(f"Ground truth: {GT_TOTAL} true columns | IOU≥{IOU_THRESH} | conf≥{CONF_THRESH}")
print(f"{'='*100}")
hdr = f"{'Model':<42} {'GB':>4} {'Score':>6} {'Dets':>5} {'Prec':>6} {'Rec':>6} {'F1':>6} {'Conf':>6} {'Time(s)':>8} {'Status'}"
print(hdr)
print("-"*100)
for e in display_entries:
    name  = e["model"][:41]
    gb    = f"{e['approx_gb']:.1f}" if e.get("approx_gb") else "cloud"
    score = str(e.get("score", "?"))
    dets  = str(e.get("total_columns", "?"))
    prec  = f"{e['precision']:.3f}" if e.get("precision") is not None else " N/A"
    rec   = f"{e['recall']:.3f}"    if e.get("recall")    is not None else " N/A"
    f1    = f"{e['f1']:.3f}"        if e.get("f1")        is not None else " N/A"
    conf  = f"{e['avg_confidence']:.3f}" if e.get("avg_confidence") else " N/A"
    t     = f"{e['response_time_seconds']:.0f}" if e.get("response_time_seconds") else " N/A"
    status = e.get("status", "?")
    flag  = " ← BEST" if e == display_entries[0] and isinstance(e.get("score"), int) else ""
    print(f"{name:<42} {gb:>4} {score:>6} {dets:>5} {prec:>6} {rec:>6} {f1:>6} {conf:>6} {t:>8}  {status}{flag}")

print(f"{'='*100}")

# Best model
passing = [e for e in all_entries if isinstance(e.get("score"), int) and e.get("score", 0) >= PASS_SCORE]
if passing:
    best = passing[0]
    print(f"\n✓ RECOMMENDED: {best['model']}")
    print(f"  Score {best['score']}/6 | P={best['precision']:.3f} R={best['recall']:.3f} F1={best['f1']:.3f}")
else:
    best_any = max(all_entries, key=lambda x: x.get("f1") or 0) if all_entries else None
    if best_any:
        print(f"\n⚠ No model passed threshold {PASS_SCORE}/6.")
        print(f"  Best available: {best_any['model']} — Score {best_any.get('score',0)}/6, F1={best_any.get('f1',0):.3f}")
```

## 7 — Visual Comparison

```python
tested = [e for e in all_entries
          if e.get("status") not in ("pull_failed", "detection_failed", "error")]

if not tested:
    print("No successful detections to visualise yet.")
else:
    # Bar chart: F1 scores
    names  = [e["model"].split("/")[-1][:28] for e in tested]
    f1s    = [e.get("f1") or 0.0 for e in tested]
    precs  = [e.get("precision") or 0.0 for e in tested]
    recs   = [e.get("recall") or 0.0 for e in tested]
    scores = [e.get("score", 0) for e in tested]

    x = np.arange(len(names))
    w = 0.25

    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

    ax1.bar(x - w, precs,  w, label="Precision", color="steelblue",   alpha=0.85)
    ax1.bar(x,     recs,   w, label="Recall",    color="darkorange",  alpha=0.85)
    ax1.bar(x + w, f1s,    w, label="F1",        color="forestgreen", alpha=0.85)
    ax1.axhline(0.25, color="red", linestyle="--", alpha=0.5, label="F1 threshold")
    ax1.set_xticks(x); ax1.set_xticklabels(names, rotation=25, ha="right", fontsize=8)
    ax1.set_ylabel("Score")
    ax1.set_title("Precision / Recall / F1 per Model")
    ax1.legend(fontsize=8); ax1.grid(axis="y", alpha=0.3)

    colours = ["gold" if s >= PASS_SCORE else "lightcoral" for s in scores]
    ax2.barh(names[::-1], scores[::-1], color=colours[::-1], alpha=0.9)
    ax2.axvline(PASS_SCORE, color="red", linestyle="--", label=f"Pass threshold ({PASS_SCORE})")
    for i, (s, n) in enumerate(zip(scores[::-1], names[::-1])):
        ax2.text(s + 0.05, i, str(s), va="center", fontsize=9)
    ax2.set_xlabel("Score (out of 6)")
    ax2.set_title("6-Point Rubric Score per Model")
    ax2.legend(fontsize=8); ax2.grid(axis="x", alpha=0.3)

    plt.tight_layout()
    plt.savefig("eval2_model_scores.png", dpi=120)
    plt.show()
```

```python
# Detection count vs expected
if tested:
    names2  = [e["model"].split("/")[-1][:30] for e in tested]
    counts  = [e.get("total_columns", 0) for e in tested]
    times   = [e.get("response_time_seconds") or 0 for e in tested]

    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 4))

    colours = ["steelblue" if 5 <= c <= 200 else "tomato" for c in counts]
    ax1.bar(names2, counts, color=colours, alpha=0.85)
    ax1.axhline(GT_TOTAL, color="green", linestyle="--", linewidth=2, label=f"GT = {GT_TOTAL}")
    ax1.axhspan(5, 200, alpha=0.05, color="green", label="Plausible range 5-200")
    ax1.set_ylabel("Columns detected")
    ax1.set_title("Detections per Model (green dash = ground truth)")
    ax1.tick_params(axis="x", rotation=25)
    for tick in ax1.get_xticklabels(): tick.set_fontsize(8)
    ax1.legend(fontsize=8); ax1.grid(axis="y", alpha=0.3)

    ax2.bar(names2, times, color="mediumpurple", alpha=0.85)
    ax2.set_ylabel("Total inference time (s)")
    ax2.set_title("Total Detection Time per Model")
    ax2.tick_params(axis="x", rotation=25)
    for tick in ax2.get_xticklabels(): tick.set_fontsize(8)
    ax2.grid(axis="y", alpha=0.3)

    plt.tight_layout()
    plt.savefig("eval2_count_time.png", dpi=120)
    plt.show()
```

```python
top2 = [e for e in tested if e.get("total_columns", 0) > 0][:2]
if not top2:
    print("No models with detections yet.")
else:
    fig, axes = plt.subplots(1, len(top2), figsize=(10*len(top2), 10))
    if len(top2) == 1: axes = [axes]

    for ax, entry in zip(axes, top2):
        model  = entry["model"]
        result = detect_pdf_ollama(TEST_PDF, model)
        annotated = agent.draw_detections(gt_img, result["detections"])
        scale = min(1.0, 2400 / max(annotated.size))
        thumb = annotated.resize(
            (int(annotated.width*scale), int(annotated.height*scale)), Image.LANCZOS
        )
        # Draw GT boxes in cyan
        draw = ImageDraw.Draw(thumb)
        for g in GT["tp"] + GT["fn"]:
            bb = [int(v*scale) for v in g["bbox"]]
            draw.rectangle(bb, outline=(0,220,220), width=2)

        ax.imshow(thumb); ax.axis("off")
        ax.set_title(
            f"{model.split('/')[-1]}\n"
            f"{entry.get('total_columns',0)} dets | "
            f"P={entry.get('precision',0):.2f} R={entry.get('recall',0):.2f} F1={entry.get('f1',0):.2f}\n"
            f"cyan = ground truth", fontsize=10
        )

    plt.tight_layout()
    plt.savefig("eval2_top2_overlay.png", dpi=100, bbox_inches="tight")
    plt.show()
```
